```python
import matplotlib.pyplot as plt

# Data tetap
v = 15          # kecepatan konstan (m/s)
t_total = 20   # detik

waktu = []
posisi = []

s = 0  # posisi awal

for t in range(t_total + 1):
    waktu.append(t)
    posisi.append(s)

    # If: peringatan jika posisi lewat 30 m
    if s > 40:
        print("Posisi melebihi 40 meter pada t =", t, "detik")

    s = s + v  # GLB: s = s + v * Δt (Δt = 1 detik)

plt.plot(waktu, posisi)
plt.title("Gerak Lurus Beraturan (GLB)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Posisi (meter)")
plt.grid(True)
plt.show()
```
